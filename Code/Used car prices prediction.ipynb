```python
df.model.value_counts()
```

Based on the list of rows per model above, We create a list of model names to be replaced:

```python
#Create a list of values to replace by 'Other' in the column 
model_val = ['T-Cross', 'Sharan', 'Scirocco', 'Beetle', 'Caddy Maxi Life', 'Arteon', 'Touran', 'Golf SV', 'Amarok','Tiguan Allspace', 'Shuttle', 'Jetta', 'CC', 'California','Caddy Life', 'Caddy', 'Caddy Maxi', 'Eos', 'Fox']
```

##### First OLS model run and Assumption check

```python
OLS(y, add_constant(X)).fit().summary()
```

As we can see in the table of result, there are several problems with the model. We will go through key indicators of the result to understand what goes wrong:

1. R-squared and Adj. R-squared: Good
    
       - R-squared signifies the “percentage variation in dependent that is explained by independent variables”. Here 86% means 86% variation of the variable 'price' is explained by other variables indicated in the middle of the table. 86% is a pretty good values. In fact, this value varies from 0 to 1. The more it is close to 1, the better the model is. However, in reality, if the value is above 0.9 it often means that the value is too perfect which leads to the fact that there are something wrong in the model. 
       - Ajd. R squared: this is the modified version of R-squared which is adjusted for the number of variables in the regression. The value is the same which is also a good sign.
       
2. Prob (F-Statistic): Good

    - This index tells the overall significance of the regression. The null hypothesis under this is “all the regression coefficients are equal to zero”. The Prob(F-statistics) depicts the probability of null hypothesis being true. As per the above results, probability is zero. This implies that overall the regressions is meaningful. 
    
    
3. AIC/BIC

    - AIC - Akaike’s Information Criteria: It penalizes the errors mode in case a new variable is added to the regression equation. It is calculated as number of parameters minus the likelihood of the overall model. A lower AIC implies a better model. It is used to compare models that have the same scale
    - BIC - Bayesian information criteria : A lower BIC implies a better model.
    
    
4. P>|t| - Pvalue: Good

    The P value is the probability of seeing a result as extreme as the one we are getting (a t value as large as ours) in a collection of random data in which the variable had no effect. A P of 5% or less is the generally accepted point at which to reject the null hypothesis.
    
    In our case, all p values are equal to 0. So, we can keep these variables in our model.
    
    
5. Prob(Omnibus): Not Good

    - One of the assumptions of OLS is that the errors are normally distributed. Omnibus test is performed in order to check this. Here, the null hypothesis is that the errors are normally distributed. Prob(Omnibus) is supposed to be close to the 1 in order for it to satisfy the OLS assumption. 
    
    In our case, the Prob(Omnibus) is 0, which implies that the OLS assumption is not satisfied. Due to this, the coefficients estimated out of it are not Best Linear Unbiased Estimators(BLUE).
    

6. Durbin-watson: Good

     This index relates to another assumption of OLS which is of homoscedasticity. This implies that the variance of errors is constant. A value between 1 to 2 is preferred. 
     
     In our case, it is ~1.4 implying that the regression results are reliable from the interpretation side of this metric.


7. Prob(Jarque-Bera):

    It is in line with the Omnibus test. It is also performed for the distribution analysis of the regression errors. It is supposed to agree with the results of Omnibus test. A large value of JB test indicates that the errors are not normally distributed.

    

8. Warning messages
    
    Per the 2nd warning message displayed, we note that there is probably a problem of multicollinearity between the variables in the model. 

Thus, based on the analysis of these key indicators, we conclude that:
    - the errors are not normally distributed. 
    - problem of multicollinearity between the variables in the model.

    - Drop column ‘mpg’ and ‘category_Tiguan’ -> Assumption - Multicollinearity: reduce number of columns whose VIF > 10

##### Second run and assumption check

Prepare a new dataframe for the second run - Similar dataframe as the first one

```python
#Create new dataframe for the second run
df_expand1 = df_expand.copy()
```

```python
#From the result of the first run, the 'mpg' model has the VIF that is above 10. Thus, we drop this column for the second run
df_expand1.drop(columns = 'mpg', inplace = True)
```

As concluded at the end of the first run, we will do in the second run:
- Drop column'mpg' and 'category_Tiguan'
- Apply the natural logarithm to the values the column 'Price', 'mileage'
- Reclean the outliers
- Standardize the new dataframe with log value
- Shuffle the dataframe

Step 1: Apply the Log e to the values the column 'Price', 'mileage'

```python
df_expand1['price']  =df_expand1['price'].map(np.log).reset_index(drop = True)
df_expand1['mileage'] = df_expand1['mileage'].map(np.log).reset_index(drop = True)
```

```python
df_expand1.head()
```

Step 2: Eliminate the outliers

```python
#Create a dataframe with aggregation calculation: quantile 25th and 75th.
stats = df_expand1[['price','mileage','tax','engine_size']].describe().transpose()
# stats = df_expand1[['price','mileage']].describe().transpose()
```

```python
#create a new column for iqr value
stats['iqr'] = stats['75%']-stats['25%']
```

```python
stats
```

```python
stats.loc['price','75%'] + stats.loc['price','iqr']*1.5
```

```python
outliers = pd.DataFrame(columns=['price','mileage','tax','engine_size'])
# outliers = pd.DataFrame(columns=['price','mileage'])
 
for col in stats.index:
    iqr = stats.at[col,'iqr']
    cutoff = iqr * 1.5
#     cutoff = iqr*1.7
    lower = stats.at[col,'25%'] - cutoff
    upper = stats.at[col,'75%'] + cutoff
    results = df_expand1[(df_expand1[col] < lower) | 
                   (df_expand1[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)
```

```python
#Create a list of index of the outliers so that we can use to drop directly the corresponding rows.
lst_outliers = outliers.index.tolist()

print(f'The number of outliers is:',len(lst_outliers))
```

```python
#Drop rows containing outliers based on the list of indexes above and store data without outliers in a new dataframe 
df_expand1_no_out = df_expand1.drop(lst_outliers).reset_index(drop = True)
```

Step 3: Standardize the new dataframe with log value


```python
# Standardize the dataframe
df_expand1_no_out_std = pd.DataFrame(StandardScaler().fit_transform(df_expand1_no_out), columns = df_expand1_no_out.columns)
```

```python
# df_expand1_no_out_std.drop(columns = 'category_Golf', inplace = True)
df_expand1_no_out_std.drop(columns = 'category_Tiguan', inplace = True)
```

```python
# df_expand1_no_out_std.drop(columns = 'transmission_Semi-Auto', inplace = True)
```

```python
#Shuffle the dataframe in order to avoid autocorrelation
df_expand1_no_out_std=df_expand1_no_out_std.sample(frac=1).reset_index(drop=True)
```

Run 2 - Model

```python
#Create the new X and Y variables for the run 3 of model
y_new=df_expand1_no_out_std['price']
X_new=df_expand1_no_out_std.drop('price', axis=1)
```

```python
#Run model
OLS(y_new, add_constant(X_new)).fit().summary()
```
